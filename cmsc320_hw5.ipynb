```python
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import pprint
import scipy as scipy
```

```python
path = "./drive/MyDrive/cmsc320/HW5/GPA_Dataset.csv"
df = pd.read_csv(path)
```

```python
a = df["What state are you from?"].value_counts()
print(a)
print(a.index)
colors = ["red","gray", "teal", "magenta", "green","blue","purple","pink","orange","brown"]
plt.bar(range(10), np.array(a), color=colors, label = a.index)
plt.title("States of origin for participants")
plt.legend()
ax = plt.gca()
ax.get_xaxis().set_visible(False)
plt.show()
```
